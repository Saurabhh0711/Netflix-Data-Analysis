### Taking into consideration the duration of each film, which director shot the most minutes?

```python
director_duration = data1.groupby('director')['duration'].sum().reset_index()
```

```python
max_duration_director = director_duration[director_duration['duration'] == director_duration['duration'].max()]
print(max_duration_director)
```

### Who is the actor who has worked in the cast with the most actors?

```python
all_actors = data1['cast'].str.split(',').sum()
from collections import Counter
actor_counts = Counter(all_actors)
```

```python
most_common_actor = actor_counts.most_common(1)[0]
```

```python
print(f"Actor with the most co-actors: {most_common_actor[0]}")
print(f"Number of co-actors: {most_common_actor[1]}")
```
